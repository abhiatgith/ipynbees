# LOGISTIC REGRESSION

### Introduction:

Logistic regression is used for classification problems where the learning algorithm classifies the dataset into different classes or categories. Hence logistic regression is used for predicting qualitative/categorical data. Logistic regression uses the sigmoid or the logit function to come up with the best fit line to classify the data into the different classes based on the probability that a particular data instance belongs to a particular class.

### Prediction:

The prediction is made using the sigmoid function as shown below where for the given input variable x, we use the below sigmoid function to give the output value. The output value denotes the probability that x belongs to the class y (hence y = 1) given the input variable x and the parameters theta. Here the threshold value is assumed to be 0.5 which means that if the output variable is more than 0.5 then x belongs to the class 1. If you notice the sigmoid function, if the value of theta transpose x is 0, then the equation evaluates to 0.5 and hence if the value of theta transpose x is greater than 0, then the equation will evaluate to a value greater than 0.5 which means that x will belong to class 1.
