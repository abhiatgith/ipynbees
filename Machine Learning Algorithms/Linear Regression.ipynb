# Linear Regression

### Introduction:

Linear Regression is the process of identifying the optimal linear (best fit line) representation of the relationship between a set of input variables (or independent variables) and the output variable (Dependent variable) in order to build a model of this representation which can then predict the value of the output variable based on the values of the input variables provided. Linear regression is use for quantitative analysis where the output variable that you are predicting is a continuous variable.

### Equation:

The relationship or the equation for linear regression based on which the model is built is y = mx+c+e where:

x is the input variable

y is the output variable

m is the slope of the best fit line

c is the intercept of the best fit line

e is the error term

Basically, we identify the best value for the parameters m and c so that the error term is reduced as much as possible which gives us the best fit line. We use a cost function (In the case of linear regression, the cost function that we use is the Squared error function or the sum of squared errors ) to represent the amount of error that the model will incur as a result of trying to identify the best fit line and then try to reduce this cost function(error in this case) to the minimal by finding the most ideal m and c values where the error value is at its minimum.

### Gradient Descent

The process of reducing the cost function is done by the gradient descent algorithm which is an algorithm that can reduce any generic function. By reducing, we mean finding the optimum values for the parameters in a function (like m and c in this case) which will reduce the cost factor of that function (which is the sum of squared errors in this case).

The exact version of the gradient descent algorithm is the “Batch” gradient descent algorithm that is used in linear regression.

The gradient descent algorithm initializes the m (theta1) and c (theta 0) to some values first. I will be using m and c instead of theta 0 and theta 1 from now as the parameter values in the explanation below.

Then, the next step is to calculate the new values of m and c and update them simultaneously as shown in the formula above. The learning rate alpha is predetermined which determines how fast (or slow) the gradient descent algorithm will converge to the local minima. The derivative draws a tangent (line) at the cost function (When plotting the cost function against the parameters m and c of the function) and then determines or calculates the slope of that tangent. If the result of the derivative is positive (which means that the slope is a positive slope) then the new values of the m and c will be lower than the previous values of m and c. If the result of the derivative is negative which means that the slope is a negative slope, then the new values of m and c will be higher than the previous values of m and c. The goal here to update the values to m and c iteratively simultaneously until we reach the global minima or the local minima at which point the values of m and c will no longer change (when the new values of m and c are calculated) and this leads to the cost (sum of squared error) of the cost function being at the minimum.
