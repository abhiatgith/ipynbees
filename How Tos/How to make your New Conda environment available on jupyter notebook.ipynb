Once you create a new Conda environment, the new environment is not available by default as an option when you try to create a new notebook from Jupyter as shown below. Below I activated the TFOD environment that I had created newly and then opened the Jupyter notebook:








## Steps:

Hence, if you want to make the new conda environment (along with all the packages installed in the new Conda Environment) available as an option to create a new notebook in jupyter, you need to follow the below steps to install the ipykernal on the new environment:

Step 1: Run the below command from the anaconda prompt as shown below:

conda install ipykernel




Click yes to proceed on the below screen:




Step 2: Run the below command from the prompt (TFOD is the display name for the environment that will appear on jupyter notebook):

ipython kernel install --name TFOD --user





After the above two steps, when you open Jupyter notebook as shown below, you can see the new conda environment as an option to create new notebooks in Jupyter notebook:
