# GIT and GitHub - Moving your project to GitHub using GIT and GitHub

## Introduction

A typical use case of Git and Github is as follows - You have developed a local project and would like to share it with the world and also allow any user to suggest changes to your project. Once you are sure of the changes that are suggested, you as an administrator of the project can then accept the suggested changes and then the changes are incorporated into the project. This is where Git and Github can help. Git is a tool that allows you to track your local project folders as git repositories. You can then move your local projects (Now Git repositories) into GitHub (https://github.com/) repositories which is the hosting service for git repositories. Once you host your local respositories as public GitHub repositories, you have enabled any other user who is signed up on Github to potentially see your project and also suggest changes by following some specfic steps.

### GIT :

Git is a free and open source distributed version control system software to manage source code history. It is designed to handle everything from small to very large projects with speed and efficiency. You can download and install the GIT software from the below link (Ensure to enable the options of intergration with Windows Explorer as given here - https://www.geeksforgeeks.org/working-on-git-bash/ given during the Git installation):

Download and Install GIT from - https://git-scm.com/download/win

Once you install GIT on your system, you can follow the below process to initialize any folder in your system to be tracked as a GIT repository. Before I go ahead with this, below is an introduction to GitHub.

### GitHub:

GitHub is the code hosting platform/service for Git repositories. Git is the tool, GitHub is the service for projects that use Git. You need to sign up for GitHub at https://github.com/ 

GitHub lets you and others work together on projects from anywhere. It provides access control and several collaboration features such as bug tracking, feature requests, task management, continuous integration and wikis for every project.

Some GitHub essentials concepts are:

Repositories
Branches
Commits
Pull Requests
Git (the version control software GitHub is built on)

#### Repository:
A GitHub repository can be used to store a development project. It can contain folders and any type of files (HTML, CSS, JavaScript, Documents, Data, Images). A GitHub repository should also include a licence file and a README file about the project. A GitHub repository can also be used to store ideas, or any resources that you want to share.

#### Branch:
A GitHub branch is used to work with different versions of a repository at the same time. By default a repository has a main branch(Until October 2020, the default branch used to called 'master' but has been since changed to main - https://github.blog/changelog/2020-10-01-the-default-branch-for-newly-created-repositories-is-now-main/). Any other branch is a copy of the main branch (as it was at a point in time). New Branches are for bug fixes and feature work separate from the master branch. When changes are ready, they can be merged into the main branch. If you make changes to the main branch while working on a new branch, these updates can be pulled in.

#### Commits:
At GitHub, changes are called commits. 

#### Pull Requests:
Pull Requests are the heart of GitHub collaboration. With a pull request you are proposing that your changes should be merged (pulled in) with the master. Pull requests show content differences, changes, additions, and subtractions in colors (green and red). As soon as you have a commit, you can open a pull request and start a discussion, even before the code is finished.

What is git remote add …?
As you probably know, git is a distributed version control system. Most operations are done locally. To communicate with the outside world, git uses what are called remotes. These are repositories other than the one on your local disk which you can push your changes into (so that other people can see them) or pull from (so that you can get others changes). The command git remote add origin https://github.com/abhiatgith/GITTEST.git creates a new remote called origin located at https://github.com/abhiatgith/GITTEST.git. Once you do this, in your push commands, you can push to origin instead of typing out the whole URL.

Once you add the remote repository (as ‘origin’)  you can see the newly added remote repository when you use the command: git remote

- After adding the remote repository as shown above, now use the git pull command to first pull the changes from the remote repository to the local repository (There is a readme file that was created in the Github repo) using the below command:

What is git push origin master?

This is a command that says "push the commits in the local branch named master to the remote named origin". Once this is executed, all the stuff that you last synchronised with origin will be sent to the remote repository and other people will be able to see them there. In simple words git push command updates the remote repository with local commits. The origin represents a remote name where the user wants to push the changes. ... The git push command basically takes two arguments: A remote name, for example, origin. A branch name, for example, master.
